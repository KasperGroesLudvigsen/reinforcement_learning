# Task 3 - Experiment # 1 - Effect of reducing observation size

The objective here is to investigate the extent to which reducing the size of the observation (the n*n area the agent can observe) influences performance

```python
import numpy as np
import matplotlib.pyplot as plt
import learning_environment as lr
import task1.britneyworld as bw
import task1.assassinworld as aw
```

## Running experiment

```python
import params.task3_experiment_3_a_params as params_A
import params.task3_experiment_3_b_params as params_B
import params.task3_experiment_3_c_params as params_C


params_a = params_A.PARAMS
params_b = params_B.PARAMS
params_c = params_C.PARAMS

params_a["experiment_name"] = "task3_experiment_X_a"
params_b["experiment_name"] = "task3_experiment_X_b"

environment_a = bw.Environment(params_a["environment_params"])
environment_b = bw.Environment(params_b["environment_params"])
environment_c = bw.Environment(params_c["environment_params"])

lr.learning_environment(params=params_a, environment=environment_a)
lr.learning_environment(params=params_b, environment=environment_b)
lr.learning_environment(params=params_c, environment=environment_c)
```

## Plotting results

```python
a_test = np.load("task3_experiment_X_atest_scores.npy")
a_test = a_test[-100:]
b_test = np.load("task3_experiment_X_btest_scores.npy")
b_test = b_test[-100:]
c_test = np.load("experiment_nametest_scores.npy")
c_test = c_test[-100:]
a_test_av = sum(a_test)/100
b_test_av = sum(b_test)/100
c_test_av = sum(c_test)/100
print("Average test score for observation size 10: {}".format(a_test_av))
print("Average test score for observation size 5: {}".format(b_test_av))
print("Average test score for observation size 0: {}".format(c_test_av))
epochs = range(len(a_test))
fig, ax = plt.subplots()
ax.scatter(epochs,a_test,label = 'Obs_size=10', s=5)
ax.scatter(epochs,b_test,label = 'Obs_size=5', s=5)
ax.scatter(epochs,c_test,label = 'Obs_size=0', s=5)
#ax.scatter(epochs,d_val,label = 'd', s=3)
ax.legend()
ax.grid(True)
plt.title('Test scores for different observation sizes')
plt.xlabel('Episode')
plt.ylabel('Episode reward')
plt.show()
```

# Task 3 - Experiment 2 - Effect of less sparse rewards

The objective here is to investigate whether redesigning the environment such that rewards are less sparse will have an impact on algorithm performance.

We did 4 runs with the following settings:
1. Pushing is rewarded = False, episodes = 2000, observation size = 10
2. Pushing is rewarded = True,  episodes = 2000, observation size = 10
3. Pushing is rewarded = True,  episodes = 4000, observation size = 10
4. Pushing is rewarded = True,  episodes = 2000, observation size = 0

Note that run # 1 is not reproducible because it was run before we altered the environment to give less sparse rewards, and rewards for pushing is not parameterisable.
